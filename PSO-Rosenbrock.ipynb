```python
from pygmo import *
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math

#verbosity
v=15
#generations
g=100
#population
popl=20
```

```python
#PSO for dimension 2

algo = algorithm(pso(gen = g))
algo.set_verbosity(v)
prob = problem(rosenbrock(2))
pop = population(prob, popl)
pop = algo.evolve(pop)

#output for 2 dimensions

gen=[]
meanLbest=[]
uda = algo.extract(pso)
for i in range(0,int(g/v)):
    elem=uda.get_log()[i]
    gen.append(elem[0])
    meanLbest.append(elem[3])

plt.plot(gen,meanLbest)
plt.xlabel('generations')
plt.ylabel('mean best fitness')
print(gen,meanLbest)
```

```python
#PSO for dimensions [2,30,50,100] & constant population

algo = algorithm(pso(gen = g))
algo.set_verbosity(v)

logs=[]
count=0
for d in [2,30,50,100]:
    prob = problem(rosenbrock(d))
    pop = population(prob, popl)
    pop = algo.evolve(pop)

    #output/log and visualization
    gen=[]
    meanLbest=[]  
    udaLog = algo.extract(pso)
    for i in range(0,int(g/v)):
        logRow=udaLog.get_log()[i]
        gen.append(logRow[0])
        meanLbest.append(logRow[3])
    
    print('Dimensions=',d,'Mean fit',np.mean(meanLbest))
    
    plt.plot(gen,meanLbest)
    plt.axhline(y=np.mean(meanLbest), linestyle='--',label='Dim=%s'%(d))
    plt.text(gen[count],meanLbest[count],'Dim=%s'%(d))
    count+=1

#visualization

plt.xlabel('generations')
plt.ylabel('mean best fitness')
```
